```python
import pandas as pd
import json
import glob

files = glob.glob("./*.json")
all_data = []
for f in files:
    with open(f,'r') as data:
        form_id = f.split("-")[-1].split(".")[0]
        form_name = f.split("-")[-2]
        data = json.load(data)
        questions = data.get("questions")
        data = data.get("questionOptions")
        for d in data:
            question = list(filter(lambda x: x["keyId"] == d["questionId"], questions))[0]
            d.update({"form_id": form_id, "form_name": form_name,"variable": question.get("variableName")})
            all_data.append(d)
data = pd.DataFrame(all_data)
data = data[data['code'] == data['code']]
csv_data = data[['text','variable','code','order','questionId','form_id','form_name']]
csv_data = csv_data.sort_values(by=['questionId'])
csv_data.to_csv("./option_with_codes.csv", index=False)
data = data[['form_id','form_name','questionId','variable','text','code']]
data = data.groupby(['form_id','form_name','questionId','variable','text','code']).first()
data.to_excel("option_list.xlsx")
```

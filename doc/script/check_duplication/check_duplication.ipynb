```python
import pandas as pd
import json
import os
```

```python
dir_path = "../../../backend/source/forms"
```

```python
files = [f for f in os.listdir(dir_path) if "prod.json" in f]
```

```python
files
```

```python
qs_duplicated = {}
opt_duplicated = {}
questions = []
options = []
for form_file in files:
    file_path = os.path.join(dir_path, form_file)
    with open(file_path, 'r') as json_file:
        data = json.load(json_file)
    questions += [{
        "form": data["id"],
        "question_id": q["id"],
        "question_name": q["name"],
        "options": q.get("options"),
        "question_group": qg["id"],
    } for qg in data["question_groups"] for q in qg["questions"]]

for q in questions:
    if q.get("options"):
        for o in q["options"]:
            options.append({
                "form": q["form"],
                "question_id": q["question_id"],
                "option_id": o.get("id"),
                "option_name": o.get("value"),
            })
    # print(qs_have_options)
qs = pd.DataFrame(questions)
op = pd.DataFrame(options)
```
